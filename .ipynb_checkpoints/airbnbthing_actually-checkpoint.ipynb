#### Introduction
A huge competitor to the hotel industry is Airbnb which is a service where people can rent out their property to others. Airbnb is growing to be one of the largest hospitality services as more people are preferring to rent out an Airbnb than stay at a hotel (https://www.investopedia.com/articles/investing/112414/airbnb-brings-sharing-economy-hotels.asp) Especially during vacation time, more people are searching for affordable places to stay. However, there are multiple factors that influence Airbnb costs such as economic events that can increase housing prices. As a result, Airbnbs must rent out their property at a higher price in order to accommodate their growing house price. Higher prices, however, can affect customers’ satisfaction especially when they expect more due to the high cost. We decided to do a study relating to this situation on Airbnbs in San Diego.
 
San Diego is known for its sunny weather and beaches which is why numerous tourists come to San Diego for vacation. Instead of staying at pricey hotels, many people opt to rent an Airbnb. As more people want to stay in San Diego, especially during the summer, the prices of San Diego Airbnbs’ rise. We wondered if there was a correlation between customer satisfaction and Airbnb prices. Thus, our goal for this project is to determine the relationship between pricing and satisfaction levels over time. We predict since San Diego has a lot of beaches and is associated with warm climates, there will be higher pricings over Summer and lower over Winter. In addition, because people expect more over time, we predict a trend of decreasing overall satisfaction.

### Import Data from df (processing for map)
Arcgis has the map and on the map we can apply layers. This is similar to Photoshop layers in which we can turn them on and off/add and subtract different layers. We can import data from our dataframes for now using the gis.content.import method. One problem is that you can only import 1000 feature points into a layer at a time and a layer can only hold up to approximately 10000 features.

Because of this, we can either cluster data first, or we can import multiple layers. We will do the latter. Below we are creating a list of layers each containing 1000 data points of longitude latitude coordinates

```python
#import collections in sizes of 1000
def make_ft_cols(df):
    temp_list = []
    i = 0
    while i < len(df):
        print("imported",i,"of",len(df))
        temp_list.append(mygis.content.import_data(df[i:i+1000]))
        i += 1000

    print("imported", len(df),"of",len(df),"rows")
    return temp_list
```

```python
#temp_list = make_ft_cols(df)
```

Since layers can each hold up to 10000 features and we have layers of 1000 features, we will begin merging the layers into a lesser amount of larger layers. We will combine 8 layers at a time, both to lower the amount of total layers, but we do not want to do 10 layers at a time since we want to avoid the possibility of a case where we accidently go over the limit due to extraneous factors(ex. 8 layers with 1000 features becomes 1 layer with 8000 features)

```python
def merge(ftlist):
    temp_ft_col = ftlist[0]
    merge_count = 0;
    for i in ftlist[1:]:
        merge_count += 1
        temp_ft_col = merge_layers(temp_ft_col,i,gis=mygis)
        print("merged",merge_count,"list(s)")
    return temp_ft_col
```

```python
#each ft_collection can only hold up to about 10,000 values. We merge in collections of 8
def merge_ft_cols(ft_col):   
    ind=0
    ft_cols_list = []
    while ind < len(ft_col):
        print(int(1 + ind/8),"collections of",int(len(ft_col)/8),"made")
        ft_cols_list.append(merge(ft_col[ind:ind+8]))
        ind+=8
    return ft_cols_list
```

## attempt#3 at identifying a relationship with price and satisfaction
From this point on we'll only work with data points with more than 5 reviews

```python
for year in list_total:
    year[:] = (df.dropna(subset = ['overall_satisfaction','price']) for df in year)
    year[:] = (df[df.reviews > 5] for df in year)
    year[:] = (df[df.price > 0] for df in year)
for years in list_total:
    for months in years:
        months['log_price'] = np.log10(months['price'])
```

This method takes a given month and finds the linear relationship between price and overall satisfaction, it returns the coefficients as a tuple

```python
def printTrends(yearIn,monthIn):
    a,b = np.polyfit(list_total[yearIn][monthIn]['log_price'],list_total[yearIn][monthIn]['overall_satisfaction'],1)

    x = np.arange(list_total[yearIn][monthIn]['log_price'].min() , list_total[yearIn][monthIn]['log_price'].max(),0.01)
    y =  a*x + b

    plt.plot(x,y)
    plt.scatter(list_total[yearIn][monthIn]['log_price'],list_total[yearIn][monthIn]['overall_satisfaction'], s=0.02)
    return (a,b)
```

Runs through all of the months and gets a list of all the coefficients and constant relations and adds it to a list. We will plot this and identify relationship with time using this. Below an example of how the distribution and lines look for the months


```python
coef=[]
const=[]
for i,y in enumerate(list_total):
    for j in range(len(y)):
        c = printTrends(i,j)
        coef.append(c[0])
        const.append(c[1])
```

Below we have the graphs of the coefficients and constants in relation to time. The coefficients have a negative correlation with time but constants have a positive correlation. Both seem to have high variance indicating low correlation

```python
plt.plot([numeric_time(x) for x in months_list],coef)
```

```python
plt.plot([numeric_time(x) for x in months_list],const)
```

```python
df_coef = pd.DataFrame(coef)
df_coef['time'] = [numeric_time(x) for x in months_list]
df_coef['const'] = const
df_coef.columns = ['coef','time','const']
```

```python
df_coef.corr()
```

We want to find a relationship of the coefficient and the constant together in comparison with time so we give it the formula 'coef:const ~ time'. Using this, we run an OLS. The variance is still very low so we still fail to reject the null hypothesis

```python
outcome,predictors = patsy.dmatrices('coef ~ time',df_coef)
mod = sm.OLS(outcome,predictors)
res = mod.fit()
print(res.summary())
```

## Conclusion
In summary, we were not able to find a relationship between pricing and satisfaction with respect to time. We could find no correlation between these two variables. In the beginning of this project, we thought there would be decreasing satisfaction levels over time as people's expectations continuously rise. Coming into this project, we worried that our dataset would not be sufficient to find an accurate correlation. The data we found online was inconsistent in that some years had more data than others. In addition, the data was taken from random days from random months throughout the past few years. Still, we thought there could be interesting observations made.

Initially, when plotting our data, we thought there might have been a relationship between the two variables after all. Interestingly enough, around December 2016, we noticed a significance increase in 0 star reviews. After further research on the possible reasons for this, we found that half of those datapoints simply had no reviews. If we take a look at the graphs we made to visualize our data, the amount of 0.0 values for overall satisfaction largely increases after December 2016. However, when we filter the data by removing values with reviews less than 5, this drastically changes and the distribution becomes more normal. We decided to filter out data with 5 or less reviews to fit other existing data. We suspect this to be beacause the sites with 0 star reviews tend not to be visited and are stuck with their 0 star reviews. On the other hand, naturally, people have a tendency to choose places with 5 star ratings. As a consequence, they end up getting even more positive reviews. Even so, this drastic change prior to 2017 and during 2017 is interesting and could potentially lead to further research.

When examing the results of the first OLS regression chart on our df_diff data, the R^2 value was very close to 0 and our P-value was much larger than we anticipated. We tried to identify the relationship between the two variables again by taking a list of all the coefficients and constant relations through all the months. When we graph these two with respect to time, we see that the coefficients have a negative correlation with time, but the constants have a positive correlation. Since both seem to have a high variance, this indicates that there is again low correlation between the two. What we can take away from these results is that there is very little to no correlation between time and pricing/satisfaction, as shown by our low R^2 value. The high P-value suggests that there is no statistically significant relationship between the two. The regression based on time did not provide any meaningful insights, nor did it account for any slight change in the variation of pricing and overall satisfaction. Considering these results, we fail to reject the null hypothesis and there does not seem to be a clear trend between time and the average relationship between overall satisfaction and pricing.


For the future, we would gather more data sets that allow us to determine if there are other variables that influece pricing and satisfaction, not just time. Instead, we could look into whether the certain types of neighborhoods play a role in determining the quality of an airbnb place. Specifically, whether more affluent neighborhoods tend to have higher ratings than surburban neighborhoods and more. Additionally, we would try to scrape more data to provide a consistent analysis of the changes in pricing/satisfaction in the past few years. Since we have found that time has little to no influence, we could broaden the topic of this project and look into more interesting, possible confounding variables and biases present in the data.
